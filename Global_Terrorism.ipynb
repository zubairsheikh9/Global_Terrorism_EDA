### <font color=blue> 2.Data Cleaning </font>

```python
## Checking the dataset columns

terror.columns
```

```python
## Iterating Each column from dataset

for x in terror.columns:
    print(x, end= ", ")
```

```python
## Selecting Important columns and storing into new variable

terror_df = terror[['iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'latitude','longitude','city','location','provstate','summary',
                'attacktype1_txt', 'targtype1_txt', 'motive', 'gname', 'weaptype1_txt', 'nkill','ransom','nreleased', 'nwound','divert','addnotes']]
```

```python
## Reviewing new dataframe

terror_df.head()
```

```python
## Renaming columns to readable format

terror_df = terror_df.rename(columns={'iyear':'Year', 'iday':'Day','imonth':'Month', 'country_txt':'Country','region_txt':'Region','latitude':'Latitude', 'longitude':'Longitude',
                          'city':'City','location':'Location', 'provstate':'Province','summary':'Summary', 'attacktype1_txt':'Attack_Type',
                          'targtype1_txt':'Target_Type','motive':'Motive', 'gname' :'Group',
                          'weaptype1_txt':'Weapon','divert':'Divert','ransom':'Ransom','nkill':'kill','nreleased':'Released', 'nwound':'Wounded', 'addnotes':'Notes' })
```

```python
## Now Reviewing Final Dataframe

terror_df.head()
```

```python
## Checking shape

terror_df.shape
```

**Imported all the important columns to main dataset which are main parameters for this analysis**

```python
## Checking Null values in dataset

terror_df.isnull().sum()
```

```python
## As null values are higher in some columns we will conver it into the percentage format for easier operation

missing = round((terror_df.isnull().sum()/len(terror_df) * 100),2)
missing
```

```python
## Showing columns which have null values more than 0

missing_values = missing[missing>0]
missing_values
```

**Graphical Representation of Missing_values**

```python
## Importing visualization Libraries

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## plot missing values

plt.figure(figsize=(10,7))
plt.title('Missing Values Percentage', fontsize=20, fontweight=5, color='red')
missing_values.plot(kind='bar', color='seagreen')
plt.ylabel('Missing Values %', fontsize=15, color='blue')
plt.xlabel('Missing Values Variables', fontsize=15, color='blue')
plt.xticks(rotation=90)
plt.show()
```

**NOTE:** <font color=red> After Visualising the null values column we get clear picture of missing values percentage which is more than 50% in many columns which will hamper our analysis and mislead in getting insights </font>

**Therefore, we are dropping missing value columns which are more then 50% to avoid hampering in our data analysis.**

```python
## Dropping columns

terror_df = terror_df.drop(['Motive','Ransom','Released','Divert','Notes'],axis=1)
```

**NOTE:** <font color=red> We are not going to drop location column as it is important factor, so we are imputing the categorcal value in **Location** Column</font>

#### Numerical Variables:

```python
## Reading dataset

terror_df.head()
```

```python
## Imputing numerical variables columns

terror_df.kill = terror_df.kill.fillna(0)    ## with zero

terror_df.Wounded = terror_df.Wounded.fillna(0)    ##with zero

terror_df.Latitude = terror_df.Latitude.fillna(terror_df.Latitude.mode()[0])     ##with mode

terror_df.Longitude = terror_df.Longitude.fillna(terror_df.Longitude.mode()[0])     ##wih mode
```

```python
## Now merging two columns in one variable

terror_df['Casualty'] = terror_df['kill'] + terror_df['Wounded']
```

#### Categorical Variables:

```python
## Checking missing Values

round((terror_df.isnull().sum()/len(terror_df) * 100),2)
```

```python
## Imputing categorical values

terror_df.City = terror_df.City.fillna(terror_df.City.mode()[0])   ## Imputed with mode

terror_df.Location = terror_df.Location.fillna(terror_df.Location.mode()[0])    ## Imputed with mode

terror_df.Province = terror_df.Province.fillna(terror_df.Province.mode()[0])      ## Imputed with mode
```

**Note:** Here the **Summary** Columns contains the date of incident which is given in dataset already with some anonymous details, so we are dropping it.

```python
terror_df = terror_df.drop(['Summary'],axis=1)
```

```python
## Finally Checking the missing values

terror_df.isnull().sum()
```

<font color=green> **Hence, we handled all NA Values and imputed all required values for further analysis** </font>

#### <font color=red> Region Wise Analysis </font>

```python
## creating tabular form

area = pd.crosstab(terror_df.Year, terror_df.Region)
area.head()
```
